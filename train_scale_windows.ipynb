```python


from ultralytics import YOLO

model = YOLO('yolov8l-pose.pt')
# model = YOLO('runs/pose/yolov8m-pose-imgsz640/weights/last.pt')

results = model.train(
    data='data/00--raw/football-field-detection.v15i.yolov8/data.yaml',
    imgsz=640,
    batch=4,
    device=0,
    epochs=500,
    patience=30,
    amp=False,
    mosaic=0.0,
    plots=True,
    save=True,
    seed=0,
    deterministic=True,
    name='yolov8l-pose-imgsz640-500ep-30pat'
)


model.cpu()
del model
del results
import gc
gc.collect()
import torch
torch.cuda.empty_cache()
torch.cuda.synchronize()
```
